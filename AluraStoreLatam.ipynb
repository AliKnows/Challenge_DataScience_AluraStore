### Importación de datos

#Visualizacion

## Precio de venta total de cada tienda

```python
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
total_ventas = [ingreso_total_t1, ingreso_total_t2, ingreso_total_t3, ingreso_total_t4]
plt.figure(figsize=(10, 6))
plt.bar(nombres_tiendas, total_ventas, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])

plt.title('Ingreso Total por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingreso Total')

for i, v in enumerate(total_ventas):
    plt.text(i, v + 5000000, f'{v:,.0f}', ha='center', va='bottom', fontsize=9)

plt.ylim(min(total_ventas) * 0.9, max(total_ventas) * 1.1) # Ajustar límites del eje y
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
promedios_total = [prome_calificacion_t1, prome_calificacion_t2, prome_calificacion_t3, prome_calificacion_t4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.figure(figsize=(10, 6))
plt.barh(nombres_tiendas, promedios_total, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])

plt.title('Calificación Promedio por Tienda')
plt.xlabel('Calificación Promedio')
plt.ylabel('Tienda')
plt.xlim(min(promedios_total) * 0.9, max(promedios_total) * 1.1) # Ajustar límites del eje x

for i, v in enumerate(promedios_total):
    plt.text(v + 0.01, i, f'{v:.2f}', ha='left', va='center', fontsize=10)

plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
